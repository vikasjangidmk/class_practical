**Aim:-** Write a Program to implement PCA on Iris Dataset.

**Objective:-** Applying PCA algorithm on iris to dataset to reduce dimensinality.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled
```

```python
pca = PCA(n_components=2)  # Reduce to 2 dimensions
df_pca = pca.fit_transform(df_scaled)
```

```python
df_pca
```

```python
df_pca = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2'])
```

```python
df_pca.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_pca['PC1'], df_pca['PC2'], c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Iris Dataset with PCA')
plt.show()
```

1. Learn How to implement pca algorithm on iris dataset.
2. Understand about PCA techniques for reduce dimensionality.
